```python
%sql
USE CATALOG main;
USE SCHEMA demo;

-- =========================
-- RAW / PROCESSED
-- =========================
-- points_raw: partitioned by siteId, so only Z-Order on tileId
OPTIMIZE points_raw;

-- processed_points_tiled_v2: partitioned by siteId, tileId, so Z-Order on other columns
OPTIMIZE processed_points_tiled_v2;

-- =========================
-- AGGREGATION
-- =========================
-- tile_stats_v2: partitioned by siteId, tileId
ANALYZE TABLE tile_stats_v2 COMPUTE STATISTICS;

-- =========================
-- SURFACE
-- =========================
-- surface_cells_v2: partitioned by siteId, tileId
OPTIMIZE surface_cells_v2;
OPTIMIZE surface_patches_v2;

-- surface_patches_v2: partitioned by siteId, tileId
OPTIMIZE surface_patches_v2
ZORDER BY (patchId, surfaceType);

-- =========================
-- FEATURES - API facing
-- =========================
-- features_water_bodies_v2: partitioned by siteId
OPTIMIZE features_water_bodies_v2
ZORDER BY (waterBodyId);

-- features_building_candidates_v2: partitioned by siteId, tileId
OPTIMIZE features_building_candidates_v2
ZORDER BY (buildingCandidateId);
```
