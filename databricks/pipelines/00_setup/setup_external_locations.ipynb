```python
%sql
-- Switch to Unity Catalog main catalog
USE CATALOG main;

USE SCHEMA demo;

-- Register existing raw points Delta data as External Table
-- Requires External Location to be created first (see Cell 2 & 3)
CREATE TABLE IF NOT EXISTS points_raw
USING DELTA
LOCATION 'abfss://raw@trimblegeospatialdemo.dfs.core.windows.net/points';
```

```python
%sql
-- Create External Location for raw point cloud data
-- Using existing storage credential: trimble_adls_credential

-- Create new external location with desired name
CREATE EXTERNAL LOCATION IF NOT EXISTS raw_container
URL 'abfss://raw@trimblegeospatialdemo.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `trimble_adls_credential`)
COMMENT 'Raw point cloud data storage';

-- Verify creation
DESCRIBE EXTERNAL LOCATION raw_container;
```
